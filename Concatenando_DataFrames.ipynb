# Combinando Dados Verticalmente com Python e Pandas

```
# Importando a única biblioteca que usarei neste tratamento
import pandas as pd
```

```
xlsx = pd.ExcelFile('Vendas.xlsx')

# Carregando as três planilhas do workbook Excel
vend_072016 = pd.read_excel(xlsx, 'Vendas072016') # Planilha 1
vend_082016 = pd.read_excel(xlsx, 'Vendas082016') # Planilha 2
vend_092016 = pd.read_excel(xlsx, 'Vendas092016') # Planilha 3
```

```
type(xlsx)
```

```
# Visualizando planilha 1 (Vendas072016), que agora é um dataframe
vend_072016
```

```
# Visualizando planilha 2 (Vendas082016), que agora é um dataframe
vend_082016
```

```
# Visualizando planilha 3 (Vendas092016), que agora é um dataframe
# Lembra que nassa planilha tinhamos um probleminha lá no PDF?

vend_092016
```

```
# Concatenando as planilhas verticalmente, analisando e resolvendo aquele 
# problema nas colunas (Cod_Produto e ID_Produto)

vend_final = pd.concat([vend_072016, vend_082016, vend_092016])
vend_final
```

```
# Renomeando coluna ID_Poroduto do dataframe vend_092016 para deixar no padrão das outras
# duas primeiras planilhas

# inplace = True >> Mantém a mudança salva no dataframe, depois se executar vend_092016
# para ver o dataframe, o nome ID_Produto já estará alterado para Cod_Produto como se
# fosse o nome padrão dado na coluna da planilha.

vend_092016.rename(columns={'ID_Produto':'Cod_Produto'}, inplace = True)
vend_092016 # imprime dataframe
```

```
# Após renomear a coluna, iremos repetir aquele procedimento de concatenação 
# para observar que o problema de nome nas colunas foi resolvido.

# Mas temos mais um problema, será que você sabe qual é? Não é necessariamente
# nos dados... Se você falou índex acertou! Veja que ele se repete...

vend_final = pd.concat([vend_072016, vend_082016, vend_092016])
vend_final # mostra dataframe com a coluna 4 renomeada
```

```
# Obtendo a quantidade de linhas no dataframe
vend_final.shape[0]
```

```
# A função range() entende que a coluna índice possui registros de 0 a 15 (-1)
range(vend_final.shape[0])
```

```
# Redefinindo um contador sequencial de acordo com a quantidade de registros dos 
# nossos dataframes.

vend_final.index = range(vend_final.shape[0])
vend_final # mostra dataframe com novo índice
```

```
# Para finalizar darei um nome para a coluna índice, chamarei de 'Id'.
# Pronto, três planilhas tratadas, concatenadas e pronta para ser submetida a
# novas análises ou entregar para quem necessita.

vend_final.columns.name = 'Id'
vend_final
```
